```python
import numpy as np
```

```python
grid_size = 4
actions = 4

agents = 2

Q_tables = [np.zeros((grid_size,grid_size,actions)) for _ in range(agents)]
```

```python
def choose_action(Q,state,epsilon=0.2):

    if np.random.rand()<epsilon:
        return np.random.randint(actions)

    return np.argmax(Q[state])
```

```python
def step(state,action):

    x,y = state

    if action==0:
        x=max(0,x-1)
    elif action==1:
        x=min(grid_size-1,x+1)
    elif action==2:
        y=max(0,y-1)
    else:
        y=min(grid_size-1,y+1)

    reward = 1 if (x,y)==(3,3) else -0.1

    return (x,y),reward
```

```python
# training
for episode in range(100):

    for agent in range(agents):

        state=(0,0)

        while state!=(3,3):

            action = choose_action(Q_tables[agent],state)

            next_state,reward = step(state,action)

            x,y = state
            nx,ny = next_state

            Q_tables[agent][x,y,action] += 0.1*(reward +
                0.9*np.max(Q_tables[agent][nx,ny]) -
                Q_tables[agent][x,y,action])

            state = next_state

    # Federated averaging of Q-tables
    avg_Q = sum(Q_tables)/agents

    for agent in range(agents):
        Q_tables[agent] = avg_Q.copy()
```

```python
print("Federated RL training complete")
```
