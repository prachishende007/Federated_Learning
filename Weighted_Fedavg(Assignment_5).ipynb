```python
import numpy as np
```

```python
# Simulated client models
client_models = [
    (np.array([0.5, 1.2]), 0.3),
    (np.array([0.8, 1.5]), 0.6),
    (np.array([0.4, 1.0]), 0.2)
]
```

```python
# Number of samples per client
client_sizes = [50, 100, 30]
```

```python
# Weighted Federated Averaging
def weighted_fedavg(client_models, client_sizes):
    total_samples = sum(client_sizes)

    new_W = np.zeros_like(client_models[0][0])
    new_b = 0

    for (W, b), size in zip(client_models, client_sizes):
        weight = size / total_samples
        new_W += weight * W
        new_b += weight * b

    return new_W, new_b

global_W, global_b = weighted_fedavg(client_models, client_sizes)

print("Aggregated Global Model:")
print("Global Weights:", global_W)
print("Global Bias:", global_b)

print("\nGlobal model redistributed to clients.")
```
