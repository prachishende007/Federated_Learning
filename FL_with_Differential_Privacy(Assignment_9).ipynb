```python
import numpy as np
```

```python
def add_noise(weights,noise_level=0.05):

    noise = np.random.normal(0,noise_level,weights.shape)

    return weights + noise
```

```python
def local_train_dp(data,W,b):

    X = data[:,:-1]
    y = data[:,-1]

    W_local = W.copy()
    b_local = b

    for _ in range(20):

        pred = 1/(1+np.exp(-(np.dot(X,W_local)+b_local)))

        error = pred-y

        grad_W = np.dot(X.T,error)/len(y)
        grad_b = np.mean(error)

        W_local -= 0.1*grad_W
        b_local -= 0.1*grad_b
```
